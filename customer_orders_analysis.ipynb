```python
# Course-End Project
# Analyzing Customer Orders Using Python

# ==========================================================
# 1. STORE CUSTOMER ORDERS
# ==========================================================

# List of customer names
customers = [
    "Alice",
    "Bob",
    "Charlie",
    "Diana",
    "Ethan",
    "Fiona",
    "George",
    "Hannah"
]

# Each order stored as a tuple:
# (customer_name, product, price, category)
orders = [
    ("Alice", "Laptop", 750, "Electronics"),
    ("Alice", "T-Shirt", 25, "Clothing"),
    ("Bob", "Headphones", 60, "Electronics"),
    ("Bob", "Jeans", 45, "Clothing"),
    ("Charlie", "Blender", 80, "Home Essentials"),
    ("Charlie", "Microwave", 120, "Home Essentials"),
    ("Diana", "Smartphone", 500, "Electronics"),
    ("Diana", "Dress", 70, "Clothing"),
    ("Ethan", "Notebook", 15, "Stationery"),
    ("Ethan", "Pen Set", 10, "Stationery"),
    ("Fiona", "Coffee Maker", 95, "Home Essentials"),
    ("Fiona", "Jacket", 85, "Clothing"),
    ("George", "Tablet", 220, "Electronics"),
    ("George", "Sofa Cover", 40, "Home Essentials"),
    ("Hannah", "Camera", 300, "Electronics"),
    ("Hannah", "Skirt", 55, "Clothing"),
    ("Alice", "Mouse", 30, "Electronics"),
    ("Bob", "Lamp", 35, "Home Essentials"),
    ("Charlie", "Shirt", 30, "Clothing"),
    ("Diana", "Toaster", 45, "Home Essentials")
]

# Dictionary where keys are customer names and values are lists of ordered products
customer_products = {}

for customer, product, price, category in orders:
    if customer not in customer_products:
        customer_products[customer] = []
    customer_products[customer].append(product)

# ==========================================================
# 2. CLASSIFY PRODUCTS BY CATEGORY
# ==========================================================

# Dictionary to map each product to its respective category
product_category = {}

for customer, product, price, category in orders:
    product_category[product] = category

# Set of unique product categories
unique_categories = set(product_category.values())

# ==========================================================
# 3. ANALYZE CUSTOMER ORDERS
# ==========================================================

# Calculate total spending per customer
customer_spending = {}

for customer, product, price, category in orders:
    if customer not in customer_spending:
        customer_spending[customer] = 0
    customer_spending[customer] += price

# Classify customers based on total spending
customer_classification = {}

for customer, total in customer_spending.items():
    if total > 100:
        customer_classification[customer] = "High-Value Buyer"
    elif 50 <= total <= 100:
        customer_classification[customer] = "Moderate Buyer"
    else:
        customer_classification[customer] = "Low-Value Buyer"

# ==========================================================
# 4. GENERATE BUSINESS INSIGHTS
# ==========================================================

# Total revenue per product category
category_revenue = {}

for customer, product, price, category in orders:
    if category not in category_revenue:
        category_revenue[category] = 0
    category_revenue[category] += price

# Extract unique products from all orders using a set
unique_products = set()

for customer, product, price, category in orders:
    unique_products.add(product)

# List comprehension to find all customers who purchased electronics
electronics_customers = list({
    customer for customer, product, price, category in orders
    if category == "Electronics"
})

# Top three highest-spending customers using sorting
top_three_customers = sorted(
    customer_spending.items(),
    key=lambda item: item[1],
    reverse=True
)[:3]

# ==========================================================
# 5. ORGANIZE AND DISPLAY DATA
# ==========================================================

# Dictionary to store categories purchased by each customer
customer_categories = {}

for customer, product, price, category in orders:
    if customer not in customer_categories:
        customer_categories[customer] = set()
    customer_categories[customer].add(category)

# Customers who purchased from multiple categories
multiple_category_customers = [
    customer for customer, categories in customer_categories.items()
    if len(categories) > 1
]

# Customers who bought Electronics
electronics_buyer_set = {
    customer for customer, product, price, category in orders
    if category == "Electronics"
}

# Customers who bought Clothing
clothing_buyer_set = {
    customer for customer, product, price, category in orders
    if category == "Clothing"
}

# Common customers who bought both Electronics and Clothing
common_electronics_clothing_customers = electronics_buyer_set.intersection(clothing_buyer_set)

# ==========================================================
# DISPLAY RESULTS
# ==========================================================

print("=" * 70)
print("CUSTOMER ORDERS ANALYSIS REPORT")
print("=" * 70)

print("\n1. CUSTOMER NAMES")
print("-" * 70)
for customer in customers:
    print(customer)

print("\n2. CUSTOMER ORDER DETAILS")
print("-" * 70)
for order in orders:
    print(order)

print("\n3. CUSTOMER -> ORDERED PRODUCTS")
print("-" * 70)
for customer, products in customer_products.items():
    print(f"{customer}: {products}")

print("\n4. PRODUCT -> CATEGORY MAPPING")
print("-" * 70)
for product, category in product_category.items():
    print(f"{product}: {category}")

print("\n5. UNIQUE PRODUCT CATEGORIES")
print("-" * 70)
for category in unique_categories:
    print(category)

print("\n6. CUSTOMER SPENDING AND CLASSIFICATION")
print("-" * 70)
for customer in customer_spending:
    print(
        f"{customer}: Total Spending = ${customer_spending[customer]} | "
        f"Classification = {customer_classification[customer]}"
    )

print("\n7. TOTAL REVENUE PER PRODUCT CATEGORY")
print("-" * 70)
for category, revenue in category_revenue.items():
    print(f"{category}: ${revenue}")

print("\n8. UNIQUE PRODUCTS")
print("-" * 70)
for product in unique_products:
    print(product)

print("\n9. CUSTOMERS WHO PURCHASED ELECTRONICS")
print("-" * 70)
for customer in electronics_customers:
    print(customer)

print("\n10. TOP 3 HIGHEST-SPENDING CUSTOMERS")
print("-" * 70)
rank = 1
for customer, total in top_three_customers:
    print(f"{rank}. {customer} - ${total}")
    rank += 1

print("\n11. CUSTOMERS WHO PURCHASED FROM MULTIPLE CATEGORIES")
print("-" * 70)
for customer in multiple_category_customers:
    print(f"{customer}: {customer_categories[customer]}")

print("\n12. COMMON CUSTOMERS WHO BOUGHT BOTH ELECTRONICS AND CLOTHING")
print("-" * 70)
for customer in common_electronics_clothing_customers:
    print(customer)

# ==========================================================
# FINAL BUSINESS INSIGHTS
# ==========================================================

print("\n13. KEY BUSINESS INSIGHTS")
print("-" * 70)

# Most profitable category
most_profitable_category = max(category_revenue, key=category_revenue.get)
print(f"Most Profitable Category: {most_profitable_category} (${category_revenue[most_profitable_category]})")

# High-value customers
high_value_customers = [
    customer for customer, classification in customer_classification.items()
    if classification == "High-Value Buyer"
]
print(f"High-Value Customers: {high_value_customers}")

# Most frequently purchased category
category_count = {}
for customer, product, price, category in orders:
    if category not in category_count:
        category_count[category] = 0
    category_count[category] += 1

most_purchased_category = max(category_count, key=category_count.get)
print(f"Most Frequently Purchased Category: {most_purchased_category} ({category_count[most_purchased_category]} orders)")

print("\nReport generation completed successfully.")
print("=" * 70)
```
